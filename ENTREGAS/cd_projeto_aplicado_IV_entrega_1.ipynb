# **Introdução**

O crescimento urbano acelerado e a expansão contínua das infraestruturas de transporte têm levado a um aumento significativo no volume de tráfego nas cidades brasileiras. Esse fenômeno resulta em problemas críticos, como congestionamentos, aumento no tempo de deslocamento, maior consumo de combustível e emissões de gases poluentes, que impactam negativamente a qualidade de vida dos cidadãos e a sustentabilidade ambiental. No Brasil, a FIRJAN (2014) calculou que a Região Metropolitana de São Paulo (RMSP) perdeu RS69,8 bilhões em 2013 devido a problemas de mobilidade urbana, o equivalente a 7,8% do PIB metropolitano; e a Região Metropolitana do Rio de Janeiro (RMRJ) perdeu R$29 bilhões, ou 8,2% do PIB da região. Para realizar essas estimativas, foram incluídos nas contas dados de gastos de congestionamentos e combustíveis, além do custo do tempo perdido para os trabalhadores.
Com o objetivo de mitigar esses problemas, a previsão precisa do fluxo de tráfego se torna essencial para o planejamento urbano e para a implementação de sistemas de transporte inteligente (ITS, na sigla em inglês). A previsão de tráfego, baseada em dados coletados de sensores, permite às autoridades municipais e aos gestores tomarem decisões informadas, tais como ajustes nos semáforos, informações sobre rotas alternativas e gerenciamento de incidentes, contribuindo assim para a fluidez do tráfego e para a segurança viária.
No entanto, realizar previsões em cenários urbanos complexos, como os do Brasil, apresenta desafios significativos devido à natureza dinâmica e não-linear do fluxo de veículos, influenciada por uma multiplicidade de fatores, como condições climáticas, eventos locais e acidentes. Nesse contexto, métodos tradicionais de previsão de tráfego, como modelos estatísticos e de regressão, muitas vezes falham em capturar essas complexidades.
Estudos recentes, como os de Zhang et al. em "Spatial-Temporal Graph Attention Networks: A Deep Learning Approach for Traffic Forecasting" e Guo et al. em "Attention-based Spatial-Temporal Graph Convolutional Networks for Traffic Flow Forecasting", demonstram que abordagens baseadas em grafos podem superar significativamente métodos tradicionais, especialmente em contextos onde a conectividade e as dependências espaciais entre diferentes locais desempenham um papel crucial.
Este trabalho propõe a construção de um modelo Spatio-Temporal Graph Attention Network (ST-GAT), baseado no estudo de Zhang et al., utilizando dados de sensores de tráfego instalados em áreas urbanas do Brasil. O objetivo é demonstrar a eficácia do ST-GAT na previsão das condições de tráfego, oferecendo uma ferramenta robusta e eficiente para o gerenciamento urbano. Além disso, esta pesquisa contribui para o avanço da área de previsão de tráfego, promovendo a implementação de soluções mais inteligentes e sustentáveis nas cidades brasileiras.



# **Motivações e justificativa**

A previsão de tráfego urbano é uma área de pesquisa de extrema relevância, especialmente em grandes centros urbanos brasileiros, onde os congestionamentos e o aumento no tempo de viagem afetam diretamente a produtividade econômica e a qualidade de vida da população. Dados recentes mostram que o Brasil possui algumas das cidades mais congestionadas do mundo. Por exemplo, São Paulo, a maior cidade do país, apresenta uma média de 30% a mais de tempo gasto no trânsito em comparação com outras cidades globais, segundo o relatório da TomTom Traffic Index de 2023. Além disso, o tempo médio de deslocamento nas capitais brasileiras é de cerca de 40 minutos por trajeto, sendo que em alguns casos pode ultrapassar uma hora, de acordo com o Instituto Brasileiro de Geografia e Estatística (IBGE).
Diante desse cenário, há um interesse crescente em desenvolver soluções tecnológicas que possam melhorar a gestão do tráfego urbano. A aplicação de modelos de aprendizado de máquina apresenta um grande potencial para endereçar essa necessidade.
Os benefícios potenciais dessa solução são numerosos e significativos. Em primeiro lugar, a capacidade de prever congestionamentos com antecedência pode permitir intervenções mais rápidas e eficazes por parte das autoridades de trânsito, melhorando a fluidez nas vias e reduzindo o tempo de viagem para os motoristas. Em segundo lugar, a otimização do fluxo de tráfego pode levar a uma redução no consumo de combustível e, consequentemente, nas emissões de gases de efeito estufa, contribuindo para metas ambientais e para a promoção de cidades mais sustentáveis. Finalmente, a implementação de tais modelos avançados pode servir como um marco para o desenvolvimento de cidades inteligentes no Brasil, incentivando a inovação tecnológica e a melhoria contínua dos serviços urbanos.


# **Objetivo**

O objetivo deste trabalho é aplicar e avaliar a eficácia de dois modelos de previsão de tráfego em áreas urbanas brasileiras. O primeiro modelo é o Attention Based Spatial-Temporal Graph Convolutional Networks (ASTGCN), proposto por Guo et al, o qual se destaca por capturar dinâmicas espaciais e temporais do tráfego. O segundo modelo, Spatial-Temporal Graph Attention Networks (ST-GAT), proposto por Zhang et al, incorpora redes de atenção espacial-temporal para melhorar a precisão das previsões. Este estudo busca demonstrar como esses modelos podem fornecer previsões mais precisas, contribuindo para otimizar a gestão do tráfego e reduzir os congestionamentos em grandes centros urbanos no Brasil.


# **Descrição da base de dados**

O conjunto de dados utilizado foi desenvolvido e disponibilizado publicamente pela Autarquia de Trânsito e Transporte Urbano do Recife (CTTU). A CTTU é uma sociedade de economia mista, integrante da Administração Indireta Municipal, responsável pelo monitoramento, controle e fiscalização do trânsito e transporte na capital pernambucana. Além disso, a CTTU regula o Sistema Municipal de Transporte Público de Passageiros, que opera sob permissão ou autorização em modais como táxis, transporte complementar (linhas interbairros e alimentadoras) e transporte escolar. A autarquia também realiza estudos e projetos voltados à melhoria da mobilidade urbana, além de promover ações de educação para o trânsito, com o objetivo de aumentar a segurança viária e a eficiência dos deslocamentos.
O dataset utilizado neste estudo apresenta a quantidade de veículos registrados por intervalos de velocidade média em períodos de 15 minutos. Os dados indicam quantos veículos transitaram em determinada faixa de velocidade em cada intervalo de tempo. Por exemplo, entre 08:00 e 08:15, no equipamento 022, foram registrados 60 veículos na faixa de velocidade de 0 a 10 km/h, e 67 veículos na faixa de 10 a 20 km/h.
Os dados abrangem o período de 2016 a 2022, sendo que, para esta análise, foram selecionados apenas os registros referentes ao ano de 2022. O conjunto de dados inclui informações de 38 equipamentos localizados em pontos estratégicos da cidade do Recife, no estado de Pernambuco.
Adicionalmente, para enriquecer o dataset original e proporcionar uma melhor compreensão espacial e temporal dos registros, foram incluídos dados complementares, como a localização geográfica dos sensores, data de instalação e a velocidade máxima permitida em cada local monitorado.
