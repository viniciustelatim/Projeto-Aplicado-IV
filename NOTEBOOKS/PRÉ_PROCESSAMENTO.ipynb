#BIBLIOTECAS

```python
# @title Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
from geopy.distance import geodesic
```

# DADOS DE VEÍCULOS

```python
# @title Importando os dados
data = pd.read_csv('/content/Dataset_tráfego.csv')
df = data[['equipamento', 'total_veículos']]
```

```python
df.head()
```

```python
# @title Invertendo o DataFrame

# Cria um índice temporário para manter a ordem original
df['index'] = df.groupby('equipamento').cumcount()

# Usa pivot para inverter o DataFrame
df_pivot = df.pivot(index='index', columns='equipamento', values='total_veículos')

# Redefine o índice para o original
df_pivot = df_pivot.rename_axis(None, axis=1).reset_index(drop=True)
```

```python
df_pivot
```

```python
# @title Removendo valores nulos
df_pivot.dropna(inplace=True)
```

```python
df_pivot
```

```python
# @title Salvando o Dataset
df_pivot.to_csv('Total_Veiculos.csv', index=False)
```

#MATRIX DE ADJACÊNCIA

```python
matrix = data[['equipamento', 'latitude', 'longitude']]
```

```python
matrix
```

```python
matrix = matrix.drop_duplicates(subset='equipamento')
```

```python
matrix.reset_index(inplace=True, drop=True)
```

```python
# @title Organizando por número do equipamento
matrix['equipamento_num'] = matrix['equipamento'].str.extract('(\d+)').astype(int)
matrix = matrix.sort_values(by='equipamento_num').drop(columns='equipamento_num')
```

```python
matrix.head()
```

```python
matrix.reset_index(inplace=True, drop=True)
```

```python
# @title Calculando a distância entre equipamentos
def calculate_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).kilometers

n = len(matrix)
matrix_adjacencia = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        if i != j:
            lat1, lon1 = matrix.loc[i, 'latitude'], matrix.loc[i, 'longitude']
            lat2, lon2 = matrix.loc[j, 'latitude'], matrix.loc[j, 'longitude']
            distance = calculate_distance(lat1, lon1, lat2, lon2)
            matrix_adjacencia[i, j] = distance

df_matrix = pd.DataFrame(matrix_adjacencia, columns=matrix['equipamento'], index=matrix['equipamento'])
```

```python
df_matrix
```

```python
# @title Salvando o DataFrame
df_matrix.to_csv('Matrix_Adjacencia.csv', index=False)
```
